### Why Machine Learning Uses `w₀ + w₁x₁ + w₂x₂ + ...` Instead of `y = mx + c`

The equation

$$
y = mx + c
$$

works well for a **single input variable**. Here:

- `m` = slope (how much `y` changes when `x` changes)
- `c` = intercept (value of `y` when `x = 0`)

However, machine learning models often have **many input features**:

- House size
- Number of bedrooms
- Age of house
- Distance to city

A single slope `m` is no longer enough.

Instead, we use:

$$
\hat{y} = w_0 + w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

where:

- `w₀` = bias/intercept (same role as `c`)
- `w₁, w₂, ..., wₙ` = weights (same role as slope `m`)
- `x₁, x₂, ..., xₙ` = input features

For one feature:

$$
\hat{y} = w_0 + w_1x_1
$$

which is mathematically identical to:

$$
y = mx + c
$$

with:

$$
m = w_1,\qquad c = w_0
$$

#### Why Not Use Slope?

The concept of a single slope comes from a 2D line.

A slope is defined as:

$$
m = \frac{\Delta y}{\Delta x}
$$

and breaks down for a vertical line because:

$$
m = \frac{\Delta y}{0}
$$

(because there is no change in x $\Delta x$ is 0 )

which is undefined.

In higher dimensions (many features), there is no single "slope" anyway. Instead, each feature gets its own weight:

$$
w_1,\; w_2,\; \ldots,\; w_n
$$

This generalizes linear equations from a line to planes and hyperplanes.

## Key Idea

`y = mx + c` is a special case of linear regression with one feature.

$$
y = mx + c
\quad\Longleftrightarrow\quad
\hat{y} = w_0 + w_1x_1
$$

Machine learning uses weights because they naturally extend to any number of input features.

- If two lines are perpendicular then the multiple of slopes is -1

# Vectors in Machine Learning

A **vector** is a quantity that has both:

- **Magnitude** (how large it is)
- **Direction** (where it points)

> A vector is always a column. In ML, Every data point $X$ (independent feature) is a vector.

A vector starting from origin is called a **positional vector**.

For example, the vector

$$
\mathbf{v} =
\begin{bmatrix}
3 \\
4
\end{bmatrix}
$$

has magnitude:

$$
||\mathbf{v}|| = \sqrt{3^2 + 4^2} = 5
$$

and points in the direction $(3,4)$.

## What Does a Vector Mean in ML?

In machine learning, vectors usually represent **features**, **weights**, or **embeddings**.

A feature vector:

$$
\mathbf{x} =
\begin{bmatrix}
\text{Age} \\
\text{Income} \\
\text{Experience}
\end{bmatrix}
$$

describes a single data point.

A weight vector:

$$
\mathbf{w} =
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}
$$

describes how important each feature is to the prediction.

## Magnitude in ML

The magnitude tells us the overall "size" of a vector:

$$
||\mathbf{x}|| =
\sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}
$$

Large magnitude often means:

- Larger feature values
- Greater distance from the origin
- Stronger activation in some models

This is why features are often normalized before training.

## Direction in ML

Direction is often more important than magnitude.

Two vectors can have different lengths but point in nearly the same direction:

$$
\begin{bmatrix}
1 \\
2
\end{bmatrix}
\quad\text{and}\quad
\begin{bmatrix}
10 \\
20
\end{bmatrix}
$$

These represent similar patterns because they point in the same direction.

Many ML algorithms compare vectors using their direction rather than their size.

# L1 and L2 Norms in Machine Learning

A **norm** measures the size (magnitude) of a vector.

For a vector:

$$
\mathbf{x} =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

the most common norms are **L1** and **L2**.

---

## L1 Norm (Manhattan Norm / Taxicab Norm)

The L1 norm is the sum of the absolute values of all components:

$$
||\mathbf{x}||_1 =
|x_1| + |x_2| + \cdots + |x_n|
$$


Here $||\mathbf{x}||_1$ the 1 means L1 norm.


### Example

For:

$$
\mathbf{x} =
\begin{bmatrix}
3 \\
-4
\end{bmatrix}
$$

$$
||\mathbf{x}||_1
=
|3| + |-4|
=
3 + 4
=
7
$$

### Why is it called Manhattan Norm?

Imagine moving through city blocks in Manhattan:

- 3 blocks east
- 4 blocks north

Total distance traveled:

$$
3 + 4 = 7
$$

You cannot move diagonally.

### Used In

**Lasso Regression**

L1 regularization penalty:

$$
\lambda \sum |w_i|
$$

Effect:

- Pushes some weights exactly to zero
- Performs automatic feature selection
- Creates sparse models

---

## L2 Norm (Euclidean Norm)

Displacement means the shortest distance between two points which is calculated using eucledian distance.

The L2 norm is the square root of the sum of squared components:

$$
||\mathbf{x}||_2
=
\sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}
$$

### Example

For:

$$
\mathbf{x} =
\begin{bmatrix}
3 \\
4
\end{bmatrix}
$$

$$
||\mathbf{x}||_2
=
\sqrt{3^2 + 4^2}
=
\sqrt{25}
=
5
$$

### Why is it called Euclidean Norm?

It measures the straight-line distance between two points using the Pythagorean theorem.

### Used In

**Ridge Regression**

L2 regularization penalty:

$$
\lambda \sum w_i^2
$$

Effect:

- Shrinks weights toward zero
- Rarely makes weights exactly zero
- Keeps all features but reduces overfitting

---

## Quick Comparison

| Property | L1 Norm | L2 Norm |
|----------|---------|---------|
| Actual Name | Manhattan / Taxicab Norm | Euclidean Norm |
| Formula | $$\sum |x_i|$$ | $$\sqrt{\sum x_i^2}$$ |
| Example for (3,4) | 7 | 5 |
| Regularization | Lasso | Ridge |
| Feature Selection | Yes | No |
| Produces Sparse Model | Yes | No |
| Handles Correlated Features | Less Stable | More Stable |

(Regularization Lasso and Ridge comes into play later)

## Rule of Thumb

- Use **L1 (Lasso)** when you want the model to automatically remove unimportant features.
- Use **L2 (Ridge)** when you want to keep all features but prevent very large weights.
- Use **Elastic Net** when you want a combination of both.

## Vector Direction in Feature Space (Not Actual Space)

In Machine Learning, when we talk about a vector's **direction**, it's crucial to understand that we are referring to its direction within a **feature space**, not necessarily a physical, real-world direction. Each axis in this multi-dimensional space corresponds to a specific feature ($x_1, x_2, \ldots, x_n$).

-   **Feature Dominance**: If a vector has a large value along one axis and small (or zero) values along others, it means that the feature corresponding to that axis is dominant for that data point.

  *Example*:
  
   For a vector $\mathbf{v} = \begin{bmatrix} 0 \\ 0 \\ 10 \\ 0 \end{bmatrix}$, it indicates that only the third feature ($x_3$) is significant for this data point, while other features have little to no influence.

-   **Feature Relationships**: The relative values and signs of components within a vector can reveal relationships between features for a particular data point.

  *Example*: For a vector $\mathbf{v} = \begin{bmatrix} 0 \\ 1 \\ 0 \\ -1 \end{bmatrix}$,
  
  it suggests that the second feature ($x_2$) and the fourth feature ($x_4$) are important, but they have an opposing relationship. If $x_2$ has a positive contribution, $x_4$ has a negative one, implying an indirect proportionality or inverse relationship within the feature set for this specific instance.

  The **unit vector**, calculated as
  
   $$ \hat{\mathbf{v}} = \frac{\mathbf{v}}{||\mathbf{v}||} $$
   
   specifically captures this directional relationship by normalizing the vector's magnitude. It shows the pure direction in feature space, allowing us to compare the relative contributions and relationships of features independent of their overall scale.

# From Linear Equations to Vectors and Matrices

In machine learning, a prediction with one feature is written as:

$$
\hat{y} = w_1x_1 + w_0
$$

where:

- $x_1$ = input feature
- $w_1$ = weight
- $w_0$ = bias (intercept)

When we have multiple features, the equation becomes:

$$
\hat{y} = w_1x_1 + w_2x_2 + \cdots + w_nx_n + w_0
$$

Instead of writing every term separately, we group the inputs and weights into **vectors**:

$$
\mathbf{W} =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
,\qquad
\mathbf{X} =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

The prediction can now be written as a **dot product**:

$$
\hat{y} = \mathbf{W}^T\mathbf{X} + w_0
$$

where:

$$
\mathbf{W}^T\mathbf{X}
=
w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

## Moving to Matrices

For a dataset containing many training examples, we stack all feature vectors into a matrix:

$$
X =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{m1} & x_{m2} & \cdots & x_{mn}
\end{bmatrix}
$$

where:

- $m$ = number of samples
- $n$ = number of features

The weight vector remains:

$$
W =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
$$


## Key Idea

- A **scalar** is a single number.
- A **vector** is an ordered list of numbers.
- A **matrix** is a collection of vectors arranged in rows and columns.

Machine learning uses vectors and matrices because they allow predictions for thousands or millions of samples to be computed efficiently using linear algebra.

## Dot Product: Where Vectors Meet ML

The prediction equation:

$$
\hat{y} = \mathbf{w}^T\mathbf{x} + b
$$

uses a dot product. We use $W^T$ because we can't multiply $W . X$ directly due to matrix multiplication rules. When two vectors $\mathbf{w}$ and $\mathbf{x}$ meet, there is an angle $\theta$ between them. The dot product can be expressed as:

$$
\mathbf{w}^T\mathbf{x}
=
||\mathbf{w}||
\,||\mathbf{x}||
\cos(\theta)
$$

The value of $\cos(\theta)$ indicates the degree of similarity or alignment between these two vectors, giving us critical insights into their relationship:
- If $\theta = 0^{\circ}$ (vectors point in the same direction, regardless of magnitude), then $\cos(0^{\circ}) = 1$. The dot product is maximized, indicating strong alignment.
- If $\theta = 180^{\circ}$ (vectors point in opposite directions), then $\cos(180^{\circ}) = -1$. The dot product is minimized (most negative), indicating strong opposition.
- If $\theta = 90^{\circ}$ (vectors are perpendicular), then $\cos(90^{\circ}) = 0$. The dot product is zero, indicating no linear relationship or orthogonal vectors.

From this, we can also define cosine similarity, which is often used to measure the similarity between two vectors:

$$
\cos(\theta) = \frac{\mathbf{w}^T\mathbf{x}}{||\mathbf{w}|| \,||\mathbf{x}||}
$$

**Interpreting the Angle and Direction:**

When we consider the dot product $\mathbf{w}^T\mathbf{x}$ (often ignoring the bias $b$ for geometric intuition):

- The sign of $\cos(\theta)$ tells us about the angle between $\mathbf{w}$ and $\mathbf{x}$:
    - If $\cos(\theta)$ is **positive**, $\theta$ is an **acute angle** ($0^{\circ} \le \theta < 90^{\circ}$). The vectors are pointing generally in the same direction.
    - If $\cos(\theta)$ is **negative**, $\theta$ is an **obtuse angle** ($90^{\circ} < \theta \le 180^{\circ}$). The vectors are pointing generally in opposite directions.
    - If $\cos(\theta)$ is **zero**, $\theta$ is exactly $90^{\circ}$. The vectors are **perpendicular** (orthogonal).

In classification tasks, the weight vector $\mathbf{w}$ is often perpendicular to the decision boundary (the line or hyperplane that separates classes).

- If $\mathbf{w}^T\mathbf{x}$ is **positive**, the feature vector $\mathbf{x}$ lies on the same side as $\mathbf{w}$ (the 'positive' side) relative to the decision boundary.
- If $\mathbf{w}^T\mathbf{x}$ is **negative**, the feature vector $\mathbf{x}$ lies on the opposite side of $\mathbf{w}$ (the 'negative' side) relative to the decision boundary.
- If $\mathbf{w}^T\mathbf{x}$ is **zero**, the feature vector $\mathbf{x}$ lies directly on the decision boundary.

Based on $\cos(\theta)$:
- Same direction $\rightarrow$ large positive prediction
- Perpendicular $\rightarrow$ little influence
- Opposite direction $\rightarrow$ negative influence


#### Matrix Multiplication

In machine learning, predictions are often written as:

$$
\hat{y} = W^T X
$$

where:

$$
W =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
,\qquad
X =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

After transposing :

$$
W^T =
\begin{bmatrix}
w_1 & w_2 & \cdots & w_n
\end{bmatrix}
$$

The matrix multiplication rule is:

> **Number of columns in the first matrix must equal the number of rows in the second matrix.**

So:

$$
(1 \times n) \cdot (n \times 1)
=
(1 \times 1)
$$

which produces a single value:

$$
W^T X
=
w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

This is simply the **dot product** of the weight vector and feature vector.

### Example

$$
W^T =
\begin{bmatrix}
2 & 3
\end{bmatrix}
,\qquad
X =
\begin{bmatrix}
4 \\
5
\end{bmatrix}
$$

$$
W^T X
=
(2 \times 4) + (3 \times 5)
=
8 + 15
=
23
$$

Therefore:

$$
\hat{y} = 23
$$

## Key Idea

Matrix multiplication lets us compute:

$$
w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

compactly and efficiently, which is why almost all machine learning models are written using vectors and matrices rather than long equations.

---

## Incorporating the Bias Term ($w_0$) into the Dot Product

When you see the prediction equation written as:

$$ \hat{y} = \mathbf{W}^T\mathbf{X} $$

and previously it was:

$$ \hat{y} = w_0 + w_1x_1 + w_2x_2 + \cdots + w_nx_n $$

It might seem like the bias term $w_0$ has been omitted. However, for mathematical convenience and to simplify notation, the bias term is often incorporated directly into the weight vector and feature vector.

### How the Bias is Included:

To include $w_0$ in the dot product, we can modify the feature vector $\mathbf{X}$ by adding a constant value of $1$ as an additional feature, and then correspondingly modify the weight vector $\mathbf{W}$ to include $w_0$ as its first (or last) component.

Let's assume we augment the feature vector $\mathbf{X}$ as follows:

$$ \mathbf{X}_{augmented} = \begin{bmatrix} 1 \\ x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} $$

And correspondingly, we augment the weight vector $\mathbf{W}$:

$$ \mathbf{W}_{augmented} = \begin{bmatrix} w_0 \\ w_1 \\ w_2 \\ \vdots \\ w_n \end{bmatrix} $$

Now, the dot product of these augmented vectors:

$$ \mathbf{W}_{augmented}^T \mathbf{X}_{augmented} = w_0 \cdot 1 + w_1x_1 + w_2x_2 + \cdots + w_nx_n $$

Which is exactly the original linear equation, including the bias term $w_0$.

$$ \hat{y} = \mathbf{W}_{augmented}^T \mathbf{X}_{augmented} = w_0 + \sum_{i=1}^{n} w_ix_i $$

### Why do this?

This technique allows us to express the entire linear model, including the intercept, as a single matrix multiplication or dot product. This simplifies the mathematical formulation, especially when dealing with optimizations, derivatives, and implementing these models in code (where it can be more efficient to handle all parameters in one vector).

## Unit Vector

A vector that has a magnitude of 1

Convert any vector into a unit vector by dividing the values by the magnitude.

For a vector $\mathbf{v}$, its unit vector $\hat{\mathbf{v}}$ is given by:
$$\hat{\mathbf{v}} = \frac{\mathbf{v}}{||\mathbf{v}||}$$
where $||\mathbf{v}||$ is the magnitude (or L2 norm) of the vector $\mathbf{v}$.

## Projection Vector

The **projection of one vector onto another** essentially tells us how much of the first vector lies in the direction of the second vector. Imagine shining a light perpendicular to a line; the shadow cast on the line is the projection.

Let $\mathbf{x}$ and $\mathbf{w}$ be two vectors, and let $\mathbf{p}$ be the projection of $\mathbf{x}$ onto $\mathbf{w}$.

- The angle between $\mathbf{x}$ and $\mathbf{w}$ ($\theta$) is the same as the angle between $\mathbf{x}$ and its projection $\mathbf{p}$ onto $\mathbf{w}$.
- The projection vector $\mathbf{p}$ has the **same direction** as $\mathbf{w}$ but potentially a **different magnitude**.

From basic trigonometry, in a right-angled triangle formed by $\mathbf{x}$, $\mathbf{p}$, and the perpendicular from the tip of $\mathbf{x}$ to the line of $\mathbf{w}$:

(the weight vector $\mathbf{w}$ is perpendicular to the decision boundary (or prediction line). It defines the orientation of this boundary.)

$$
\cos(\theta) = \frac{\text{adjacent}}{\text{hypotenuse}} = \frac{||\mathbf{p}||}{||\mathbf{x}||}
$$

So, the magnitude of the projection is:

$$
||\mathbf{p}|| = ||\mathbf{x}|| \cos(\theta)
$$

We also know from the dot product section that:

$$
\cos(\theta) = \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{x}|| \,||\mathbf{w}||}
$$

Now, substitute this expression for $\cos(\theta)$ into the magnitude of projection formula:

$$
||\mathbf{p}|| = ||\mathbf{x}|| \left( \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{x}|| \,||\mathbf{w}||} \right)
$$

We can cancel out $||\mathbf{x}||$:

$$
||\mathbf{p}|| = \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{w}||}
$$


This gives us the magnitude of the projection of $\mathbf{x}$ onto $\mathbf{w}$.

To get the actual projection **vector** $\mathbf{p}$, we multiply its magnitude by the unit vector in the direction of $\mathbf{w}$ (which is $\frac{\mathbf{w}}{||\mathbf{w}||}$):

$$
\mathbf{p} = ||\mathbf{p}|| \left( \frac{\mathbf{w}}{||\mathbf{w}||} \right) = \left( \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{w}||} \right) \left( \frac{\mathbf{w}}{||\mathbf{w}||} \right)
$$

Thus, the projection vector $\mathbf{p}$ is:

$$
\mathbf{p} = \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{w}||^2} \mathbf{w}
$$

> In essence, to calculate the projection vector, we first find its magnitude by taking the magnitude of $\mathbf{x}$ multiplied by the cosine of the angle between $\mathbf{x}$ and $\mathbf{w}$. Then, we give it the correct direction by multiplying this magnitude by the unit vector of $\mathbf{w}$.

### Relevance in Machine Learning

Projection vectors are crucial in various ML contexts:
- **Dimensionality Reduction**: Techniques like Principal Component Analysis (PCA) project high-dimensional data onto lower-dimensional subspaces, effectively finding the 'shadow' of the data on important directions.
- **Feature Engineering**: Creating new features by projecting existing ones onto a specific direction that might be more informative for a model.
- **Geometric Understanding**: Helps visualize how data points relate to decision boundaries or other important vectors (like weight vectors in linear models).

### Clarification on the Projection Vector Formula

The projection vector $\mathbf{p}$ must be a vector that points in the same direction as $\mathbf{w}$. The formula we derived:

$$ \mathbf{p} = \frac{\mathbf{x}^T\mathbf{w}}{|\|\mathbf{w}\||^2} \mathbf{w} $$

is correct because:

1.  **Scalar Factor**: The term $\frac{\mathbf{x}^T\mathbf{w}}{|\|\mathbf{w}\||^2}$ is a **scalar**. It represents the magnitude of the projection along the direction of $\mathbf{w}$.
    *   $\mathbf{x}^T\mathbf{w}$ is a scalar (the dot product).
    *   $|\|\mathbf{w}\||^2$ is also a scalar (the squared magnitude of $\mathbf{w}$).
    *   A scalar divided by a scalar is still a scalar.

2.  **Direction**: This scalar factor is then multiplied by the vector $\mathbf{w}$. Multiplying a scalar by a vector results in a new vector that is scaled (longer or shorter) but retains the **original direction** of $\mathbf{w}$. This is precisely what a projection vector should do: lie along the direction of the vector it's projected onto.

### Why $\mathbf{p} = \mathbf{x}^T\mathbf{w}^2 / |\|\mathbf{w}\||^2$ is not correct:

The expression $\mathbf{w}^2$ (a vector squared) is not a standard mathematical operation in this context. While it could sometimes refer to an element-wise square or $|\|\mathbf{w}\||^2$ (the squared magnitude), if it were meant as an element-wise square of the vector $\mathbf{w}$, then multiplying it by the scalar $\mathbf{x}^T\mathbf{w}$ would result in a vector, but it would not correctly represent the projection along $\mathbf{w}$.

More critically, if you interpret $\mathbf{w}^2$ as $|\|\mathbf{w}\||^2$, then the formula would become:

$$ \mathbf{p} = \frac{\mathbf{x}^T\mathbf{w} \cdot |\|\mathbf{w}\||^2}{|\|\mathbf{w}\||^2} = \mathbf{x}^T\mathbf{w} $$

This result (which is a scalar) cannot be a projection **vector**. A projection is inherently a vector quantity, having both magnitude and direction.

## Projection Vector and Prediction: A Conceptual Link

At a high level, the **projection vector** $\mathbf{p}$ of an input feature vector $\mathbf{x}$ onto a weight vector $\mathbf{w}$ can be thought of as:

*   **How much of $\mathbf{x}$ lies in the direction of $\mathbf{w}$?**

This value, the magnitude of the projection $|\|\mathbf{p}\||$, effectively quantifies the 'alignment' or 'contribution' of the input $\mathbf{x}$ along the direction that the model (represented by $\mathbf{w}$) considers important. In essence, this alignment is what forms the basis of the prediction, $\hat{y}$. The dot product $\mathbf{x}^T\mathbf{w}$ (which is directly related to the projection's magnitude) is the core of how the model calculates its output based on the input features and their learned importance.

## Scalar Projection (Component)

The **scalar projection** of vector $\mathbf{x}$ onto vector $\mathbf{w}$ is a single number (a scalar) that represents the signed magnitude of the vector projection $\mathbf{p}$. It tells us 'how much' of $\mathbf{x}$ is aligned with $\mathbf{w}$, but without specifying the direction (as the vector projection does).

### Formula:

The scalar projection of $\mathbf{x}$ onto $\mathbf{w}$ is denoted as $\text{comp}_{\mathbf{w}}\mathbf{x}$ or $p_w$ and can be calculated using either:

1.  **Using the angle:**
    $$ \text{comp}_{\mathbf{w}}\mathbf{x} = |\|\mathbf{x}\|| \cos(\theta) $$

2.  **Using the dot product:**
    $$ \text{comp}_{\mathbf{w}}\mathbf{x} = \frac{\mathbf{x}^T\mathbf{w}}{|\|\mathbf{w}\||} $$

Notice that this is simply the magnitude of the vector projection $\mathbf{p}$.

### Where it's used in ML:

Scalar projection is useful when:

-   **Quantifying influence**: You want to know the extent to which one feature vector influences another direction (e.g., how much an input feature vector aligns with a specific weight vector's learned direction), without needing the full vector result.
-   **Feature extraction**: In some cases, a new feature can be derived by projecting existing features onto a particular direction and taking only the scalar component.
-   **Interpreting dot products**: The dot product $\mathbf{x}^T\mathbf{w}$ is directly proportional to the scalar projection. Understanding the scalar projection helps interpret the dot product's meaning in terms of alignment and contribution.

## Vectors Representing Words: An Intuitive Example

In Machine Learning, not just numerical data but even abstract concepts like words can be represented as vectors. These are often called **word embeddings**.

Imagine a multi-dimensional space where each dimension represents a certain characteristic or semantic property. Words that are similar in meaning or context will be positioned close to each other in this space.

Let's take your example of "King" and "man":

Consider a simplified 2-dimensional space with dimensions like `Royalty` and `Gender`.

- **Man**: Might have a vector like $\begin{bmatrix} 0.2 \\ 0.9 \end{bmatrix}$ (low royalty, high masculinity).
- **Woman**: Might have a vector like $\begin{bmatrix} 0.1 \\ -0.9 \end{bmatrix}$ (low royalty, high femininity).
- **King**: Might have a vector like $\begin{bmatrix} 0.9 \\ 0.8 \end{bmatrix}$ (high royalty, high masculinity).
- **Queen**: Might have a vector like $\begin{bmatrix} 0.8 \\ -0.7 \end{bmatrix}$ (high royalty, high femininity).

When we look at the vectors for "King" and "man":

- **King vector** = $\begin{bmatrix} 0.9 \\ 0.8 \end{bmatrix}$
- **Man vector** = $\begin{bmatrix} 0.2 \\ 0.9 \end{bmatrix}$

Even though they are not identical, they share a positive value in the 'Gender' dimension (representing masculinity) and are generally in the same quadrant of the space, indicating a conceptual similarity. They are closer to each other than "King" is to "Woman" or "Queen" is to "Man" based on gender aspects.

### How Similarity is Measured

The similarity between these word vectors is often measured using **cosine similarity**, which is derived from the dot product. As we discussed, cosine similarity measures the cosine of the angle between two vectors.

- If two vectors are very similar (point in nearly the same direction), the angle between them is small, and $\cos(\theta)$ will be close to 1.
- If two vectors are dissimilar (point in different directions), the angle is large, and $\cos(\theta)$ will be closer to 0 or even -1.

For "King" and "man", their vectors will have a high cosine similarity score because they align well in the vector space, reflecting their semantic relationship.

This demonstrates how vectors allow computers to understand and process human language by representing words as numerical entities that can be mathematically compared and manipulated.



## Fundamental Notations: Scalars and Vectors

It's important to be clear about the notation used for single numbers (scalars) and collections of numbers (vectors).

### Scalar vs. Vector Notation

-   **Scalar (Single Value)**:
    A single number, typically denoted by a lowercase letter like $x$.
    Sometimes, to indicate it's an average, it might be written with a bar over it: $\bar{x}$ (read as "x bar"). For example, $\bar{x}$ often represents the mean of a set of values.

    *Example: The temperature $x = 25^{\circ}C$, or the average score $\bar{x} = 85$.*

-   **Vector (Ordered List of Values)**:
    A collection of numbers arranged in a specific order, typically denoted by a lowercase bold letter like $\mathbf{x}$ or a letter with an arrow above it: $\vec{x}$ (read as "x vector").
    In machine learning, bold lowercase letters (e.g., $\mathbf{x}$, $\mathbf{w}$) are the most common convention for vectors.

    *Example: A 2D position vector $\mathbf{x} = \begin{bmatrix} 3 \\ 4 \end{bmatrix}$, or a 3D feature vector $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}$.*

    **Key Distinction**: While $\bar{x}$ denotes a scalar (an average), $\vec{x}$ or $\mathbf{x}$ denotes a vector (an ordered list of components).

  The symbol $\mathbb{R}$ (or just $R$) represents the set of all real numbers, which includes all rational and irrational numbers. When you see:
  $x \in \mathbb{R}$

  *Example: $x = 5$, $x = -3.14$, $x = \sqrt{2}$ are all in $\mathbb{R}$.*

- **n-dimensional Real Space ($\,\mathbb{R}^n$ or $R^n$)**:

  The notation $\mathbb{R}^n$ (or $R^n$) represents an $n$-dimensional space where each dimension is a real number. When you see:
  $\mathbf{x} \in \mathbb{R}^n$
  This means "the vector $\mathbf{x}$ belongs to the $n$-dimensional real space," indicating that $\mathbf{x}$ is a vector with $n$ real-valued components.

  Example: $\mathbf{x} \in \mathbb{R}^2$ means $\mathbf{x}$ is a 2-dimensional vector with real components, like $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ where $x_1 \in \mathbb{R}$ and $x_2 \in \mathbb{R}$.


  Example: $\mathbf{x} \in \mathbb{R}^3$ means $\mathbf{x}$ is a 3-dimensional vector with real components, like $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}$.

These notations are fundamental for precisely describing the nature and dimensionality of variables in machine learning.

## Shifting the Decision Boundary (Line/Hyperplane)

In machine learning, a linear model's decision boundary is often represented by an equation like:

$$ w_1x_1 + w_2x_2 + w_0 = 0 $$

This equation defines a line (in 2D) or a hyperplane (in higher dimensions) that separates different classes. We can shift this boundary by modifying the parameters ($w_0, w_1, w_2$).

### Horizontal Shift (along $x_1$ axis, effectively changing the intercept)

If we want to shift the line by a constant `a` along the $x_1$ axis such that the new $x_1$ becomes $x_1' = x_1 + a$, we can see its effect on the original equation:

Original point: $(x_1, x_2)$
New point for shifted line: $(x_1-a, x_2)$ (because we want the new line to pass through the old point plus 'a' shift).

Let's consider the effect of adding `a` to $x_1$ in the equation for a new point $(x_1', x_2)$ that lies on the original line. If we want the *line itself* to shift such that the old $(x_1, x_2)$ point on the old line is now $(x_1+a, x_2)$ on the new line, then we replace $x_1$ with $(x_1-a)$ in the equation to describe the new line:

$$ w_1(x_1 - a) + w_2x_2 + w_0 = 0 $$

Expanding this:

$$ w_1x_1 - w_1a + w_2x_2 + w_0 = 0 $$

Rearranging terms:

$$ w_1x_1 + w_2x_2 + (w_0 - w_1a) = 0 $$

Here, `w₁a` is a constant. So, the new bias term `w₀'` becomes:

$$ w_0' = w_0 - w_1a $$

Thus, the shifted line is:

$$ w_1x_1 + w_2x_2 + w_0' = 0 $$

This shows that **horizontal shifts (left or right)** of the decision boundary can be achieved simply by modifying the intercept (bias term) `w₀`. To move the line right, we decrease `w₀`, and to move it left, we increase `w₀` (assuming $w_1 > 0$).

### Vertical Shift (along $x_2$ axis)

Similarly, to shift the line **vertically (up or down)** by a constant `b` (i.e., $x_2' = x_2 + b$), we replace $x_2$ with $(x_2 - b)$ in the original equation:

$$ w_1x_1 + w_2(x_2 - b) + w_0 = 0 $$

Expanding this:

$$ w_1x_1 + w_2x_2 - w_2b + w_0 = 0 $$

Rearranging terms:

$$ w_1x_1 + w_2x_2 + (w_0 - w_2b) = 0 $$

Again, the shift is absorbed into a modified bias term: $w_0'' = w_0 - w_2b$.

In both cases, shifting the line (or hyperplane) in a parallel manner effectively changes only the bias term ($w_0$), while the weights ($w_1, w_2$) dictate the orientation (slope) of the boundary.

## Orthogonality and Decision Boundaries: The Normal Vector

One of the most fundamental geometric properties in linear machine learning models (like logistic regression or Support Vector Machines) is the relationship between the **weight vector** ($\mathbf{w}$) and the **decision boundary**.

### The Weight Vector as a Normal Vector

In a linear classification model, the decision boundary is defined by the equation:

$$ \mathbf{w}^T\mathbf{x} + w_0 = 0 $$

When we consider points that lie *exactly on* this boundary, the term $\mathbf{w}^T\mathbf{x}$ (if we absorb $w_0$ into $\mathbf{w}$ and $\mathbf{x}$ as discussed in "Incorporating the Bias Term") or more generally, for any two points $\mathbf{x}_1$ and $\mathbf{x}_2$ on the boundary, their difference $(\mathbf{x}_1 - \mathbf{x}_2)$ lies *within* the boundary hyperplane.

It is a crucial geometric property that the **weight vector $\mathbf{w}$ is always perpendicular (orthogonal) to the decision boundary (line or hyperplane)**. For this reason, $\mathbf{w}$ is often referred to as the **normal vector** to the decision boundary.

This means that the angle between the weight vector $\mathbf{w}$ and any vector lying *within* the decision boundary will be $90^{\circ}$.

### Mathematical Explanation

Consider the simplified case where the decision boundary passes through the origin (i.e., $w_0 = 0$). The equation becomes:

$$ \mathbf{w}^T\mathbf{x} = 0 $$

This equation means that for any point $\mathbf{x}$ that lies on the decision boundary, its dot product with the weight vector $\mathbf{w}$ is zero. As established earlier, a dot product of zero implies that the two vectors are **orthogonal** (perpendicular).

### Practical Use and Interpretation in ML

1.  **Defining Orientation**: The direction of the weight vector $\mathbf{w}$ directly determines the orientation of the decision boundary. If you rotate $\mathbf{w}$, the decision boundary also rotates.

2.  **Class Separation**: The weight vector points in the direction where the model's output values increase most rapidly. Therefore, if $\mathbf{w}$ points towards one class (e.g., the positive class), then data points on the 'other side' (the negative class) will have their feature vectors pointing in a direction generally opposite to $\mathbf{w}$.

3.  **Model Interpretation**: Understanding that $\mathbf{w}$ is orthogonal to the decision boundary helps visualize how the model separates data points. The magnitude of $\mathbf{w}$ influences the 'strength' or 'confidence' of the separation, while its direction defines the separation itself.

4.  **Distance to Boundary**: The (signed) distance of a data point $\mathbf{x}$ from the decision boundary is proportional to $\mathbf{w}^T\mathbf{x} + w_0$. This concept is central to algorithms like Support Vector Machines (SVMs), where the goal is to maximize this distance for the closest points (support vectors).

## Distance of a Decision Boundary (Hyperplane) from the Origin

Let's derive the formula for the shortest distance `d` from the origin to a linear decision boundary (or hyperplane) defined by the equation:

$$ \mathbf{w}^T\mathbf{x} + w_0 = 0 $$

Where:
- $\mathbf{w}$ is the weight vector (normal to the hyperplane).
- $\mathbf{x}$ is a point on the hyperplane.
- $w_0$ is the bias term.

### Derivation:

1.  **Consider a point on the boundary**: Let $\mathbf{x}_0$ be any point that lies *on* the decision boundary. By definition, this point satisfies the boundary equation:

    $$ \mathbf{w}^T\mathbf{x}_0 + w_0 = 0 $$

    From this, we can express the term $\mathbf{w}^T\mathbf{x}_0$ as:

    $$ \mathbf{w}^T\mathbf{x}_0 = -w_0 \quad (1) $$

2.  **Geometric Interpretation**: The shortest distance `d` from the origin to the hyperplane is along the direction of the normal vector $\mathbf{w}$. If $\theta$ is the angle between the vector $\mathbf{x}_0$ (pointing from the origin to a point on the hyperplane) and the normal vector $\mathbf{w}$, then the distance `d` can be found using trigonometry:

    $$ d = ||\mathbf{x}_0|| \cos(\theta) $$

    (Imagine a right-angled triangle where the hypotenuse is $||\mathbf{x}_0||$ and the adjacent side is $d$, with $\mathbf{w}$ defining the direction of $d$).

3.  **Using the Dot Product Definition**: We know that the dot product of two vectors can also be expressed in terms of their magnitudes and the cosine of the angle between them:

    $$ \mathbf{w}^T\mathbf{x}_0 = ||\mathbf{w}|| ||\mathbf{x}_0|| \cos(\theta) $$

4.  **Substitute and Solve for d**: We can see that $||\mathbf{x}_0|| \cos(\theta)$ is exactly `d` from our geometric interpretation. So, we can substitute `d` into the dot product equation:

    $$ \mathbf{w}^T\mathbf{x}_0 = ||\mathbf{w}|| \cdot d $$

5.  **Combine with (1)**: Now, we have two expressions for $\mathbf{w}^T\mathbf{x}_0$. Let's equate them:

    $$ -w_0 = ||\mathbf{w}|| \cdot d $$

6.  **Final Formula for d**: Solving for `d`:

    $$ d = \frac{-w_0}{||\mathbf{w}||} $$

    This formula gives the signed distance of the hyperplane from the origin. The sign indicates which side of the origin the hyperplane lies relative to the direction of $\mathbf{w}$. If we are interested in just the magnitude of the distance, we take the absolute value:

    $$ d = \frac{|w_0|}{||\mathbf{w}||} $$

## Distance of a Data Point from a Decision Boundary (Hyperplane)

Now, let's derive the formula for the shortest distance `d` from an arbitrary data point $\mathbf{x}_0$ to a linear decision boundary (or hyperplane) defined by the equation:

$$ \mathbf{w}^T\mathbf{x} + w_0 = 0 $$

Where:
- $\mathbf{w}$ is the weight vector (normal to the hyperplane).
- $\mathbf{x}$ is a point on the hyperplane.
- $w_0$ is the bias term.

### Derivation:

1.  **Recall Distance from Origin**: We previously derived the signed distance from the origin to the hyperplane as:

    $$ d_{origin} = \frac{-w_0}{||\mathbf{w}||} \quad (1) $$

2.  **Translate the Origin**: To find the distance from an arbitrary point $\mathbf{x}_0$ to the hyperplane, we can conceptually 'shift' our coordinate system so that $\mathbf{x}_0$ becomes the new origin. In this new coordinate system, any point $\mathbf{x}'$ relative to $\mathbf{x}_0$ is given by $\mathbf{x}' = \mathbf{x} - \mathbf{x}_0$.

    Substitute $\mathbf{x} = \mathbf{x}' + \mathbf{x}_0$ into the hyperplane equation:

    $$ \mathbf{w}^T(\mathbf{x}' + \mathbf{x}_0) + w_0 = 0 $$
    $$ \mathbf{w}^T\mathbf{x}' + \mathbf{w}^T\mathbf{x}_0 + w_0 = 0 $$

    This equation describes the same hyperplane, but now with $\mathbf{x}'$ as the variable, and the origin for $\mathbf{x}'$ is at our original point $\mathbf{x}_0$.

3.  **New 'Bias' Term**: In this shifted coordinate system, the new effective bias term is $(\mathbf{w}^T\mathbf{x}_0 + w_0)$.

4.  **Apply Distance from Origin Formula**: Now, using the formula from step (1) for this new setup (distance from the *new* origin $\mathbf{x}_0$ to the hyperplane), the signed distance `d` is:

    $$ d = \frac{-(\mathbf{w}^T\mathbf{x}_0 + w_0)}{||\mathbf{w}||} \quad (Incorrect sign, let's re-evaluate) $$

    Let's think about this differently using projection, which is more intuitive.

### Alternative Derivation using Projection:

Consider a data point $\mathbf{x}_0$. Let $\mathbf{x}_p$ be the point on the hyperplane that is closest to $\mathbf{x}_0$. The vector $(\mathbf{x}_0 - \mathbf{x}_p)$ is perpendicular to the hyperplane, meaning it is parallel to the normal vector $\mathbf{w}$.

The distance `d` is the magnitude of this vector, and its direction is either same or opposite to $\mathbf{w}$. Thus, `d` can be found by projecting the vector from any point on the plane to $\mathbf{x}_0$ onto the normal vector $\mathbf{w}$.

Let $\mathbf{x}_{any}$ be any point on the hyperplane. Then $\mathbf{w}^T\mathbf{x}_{any} + w_0 = 0$.

The vector from $\mathbf{x}_{any}$ to $\mathbf{x}_0$ is $(\mathbf{x}_0 - \mathbf{x}_{any})$.

The scalar projection of $(\mathbf{x}_0 - \mathbf{x}_{any})$ onto $\mathbf{w}$ gives the signed distance `d`:

$$ d = \frac{(\mathbf{x}_0 - \mathbf{x}_{any})^T \mathbf{w}}{||\mathbf{w}||} $$

$$ d = \frac{\mathbf{x}_0^T\mathbf{w} - \mathbf{x}_{any}^T\mathbf{w}}{||\mathbf{w}||} $$

Since $\mathbf{x}_{any}$ is on the hyperplane, we know $\mathbf{w}^T\mathbf{x}_{any} = -w_0$. Substituting this into the equation:

$$ d = \frac{\mathbf{x}_0^T\mathbf{w} - (-w_0)}{||\mathbf{w}||} $$

$$ d = \frac{\mathbf{w}^T\mathbf{x}_0 + w_0}{||\mathbf{w}||} $$

This is the **signed distance** of the point $\mathbf{x}_0$ from the hyperplane.

### Interpretation of the Signed Distance:

The sign of `d` tells us which side of the decision boundary the point $\mathbf{x}_0$ lies on, relative to the direction of the normal vector $\mathbf{w}$:

-   **If $d > 0$**: The point $\mathbf{x}_0$ lies on the "positive" side of the hyperplane, in the same direction as the normal vector $\mathbf{w}$. This typically corresponds to one class (e.g., the positive class).

-   **If $d < 0$**: The point $\mathbf{x}_0$ lies on the "negative" side of the hyperplane, in the opposite direction to the normal vector $\mathbf{w}$. This typically corresponds to the other class (e.g., the negative class).

-   **If $d = 0$**: The point $\mathbf{x}_0$ lies exactly *on* the decision boundary (hyperplane).

## Distance Between Two Parallel Hyperplanes

Consider two parallel hyperplanes defined by the equations:

1.  Hyperplane 1: $H_1: \mathbf{w}^T\mathbf{x} + w_{0,1} = 0$
2.  Hyperplane 2: $H_2: \mathbf{w}^T\mathbf{x} + w_{0,2} = 0$

For these hyperplanes to be parallel, they must share the same normal vector $\mathbf{w}$ but have different bias terms $w_{0,1}$ and $w_{0,2}$.

### Derivation:

We can find the distance between these two parallel hyperplanes by utilizing our previous derivation for the signed distance of a hyperplane from the origin.

From the "Distance of a Decision Boundary (Hyperplane) from the Origin" section, we know the signed distance `d` from the origin to a hyperplane $\mathbf{w}^T\mathbf{x} + w_0 = 0$ is:

$$ d = \frac{-w_0}{||\mathbf{w}||} $$

Applying this to our two parallel hyperplanes:

-   The signed distance of $H_1$ from the origin is: $d_1 = \frac{-w_{0,1}}{||\mathbf{w}||}$
-   The signed distance of $H_2$ from the origin is: $d_2 = \frac{-w_{0,2}}{||\mathbf{w}||}$

The distance between the two parallel hyperplanes is the absolute difference between their signed distances from the origin:

$$ \text{Distance}(H_1, H_2) = |d_2 - d_1| $$

Substitute the expressions for $d_1$ and $d_2$:

$$ \text{Distance}(H_1, H_2) = \left| \frac{-w_{0,2}}{||\mathbf{w}||} - \left( \frac{-w_{0,1}}{||\mathbf{w}||} \right) \right| $$

$$ \text{Distance}(H_1, H_2) = \left| \frac{-w_{0,2} + w_{0,1}}{||\mathbf{w}||} \right| $$

$$ \text{Distance}(H_1, H_2) = \left| \frac{w_{0,1} - w_{0,2}}{||\mathbf{w}||} \right| $$

Or equivalently, since $|A - B| = |B - A|$:

$$ \text{Distance}(H_1, H_2) = \frac{|w_{0,2} - w_{0,1}|}{||\mathbf{w}||} $$

This formula gives the shortest (perpendicular) distance between two parallel hyperplanes.

## Distance Between Non-Parallel Hyperplanes

If two hyperplanes (or lines in 2D, planes in 3D) are **not parallel**, it means they are oriented differently in space. This difference in orientation guarantees that they will **intersect** at some point or along some higher-dimensional region.

### Explanation:

-   **Intersection**: When two non-parallel hyperplanes intersect, they share one or more common points. The set of all such common points forms their intersection.

-   **Shortest Distance**: The shortest distance between any two geometric objects is defined as the minimum distance between any point on the first object and any point on the second object. If the objects intersect, there exist points that are common to both. For any such common point, the distance from that point to itself (on the other object) is zero.

Therefore, for any two non-parallel hyperplanes, since they intersect, the shortest distance between them is always **0**.

## Non-Perpendicular Vectors and Angle Interpretation

We've discussed that if two vectors are perpendicular, their dot product is zero, implying no linear correlation. However, in many ML scenarios, vectors are not perpendicular. Understanding the angle between them provides crucial insights.

Recall the dot product formula:

$$ \mathbf{a}^T\mathbf{b} = ||\mathbf{a}|| \,||\mathbf{b}|| \cos(\theta) $$

Where $\theta$ is the angle between vectors $\mathbf{a}$ and $\mathbf{b}$.

### Cases when Vectors are Not Perpendicular:

1.  **Acute Angle ($\mathbf{0^{\circ} \le \theta < 90^{\circ}}$)**:
    *   **$\cos(\theta) > 0$**: When the angle between two vectors is acute, the cosine of the angle is positive. This means their dot product (and thus their projection onto each other) will be positive.
    *   **Interpretation in ML**: This indicates that the vectors are pointing in generally the same direction. In the context of feature vectors and weight vectors, a positive dot product suggests that the feature vector aligns with the positive direction learned by the weight vector, often implying that the data point belongs to the 'positive' class or contributes positively to the prediction.

2.  **Obtuse Angle ($\mathbf{90^{\circ} < \theta \le 180^{\circ}}$)**:
    *   **$\cos(\theta) < 0$**: When the angle between two vectors is obtuse, the cosine of the angle is negative. Consequently, their dot product (and projection) will be negative.
    *   **Interpretation in ML**: This indicates that the vectors are pointing in generally opposite directions. A negative dot product often means the feature vector aligns with the negative direction learned by the weight vector, suggesting the data point might belong to the 'negative' class or contributes negatively to the prediction.

### Summary:

-   **Acute angle** ($\cos(\theta) > 0$): Vectors are generally aligned, positive correlation/contribution.
-   **Obtuse angle** ($\cos(\theta) < 0$): Vectors are generally opposed, negative correlation/contribution.
-   **Right angle** ($\cos(\theta) = 0$): Vectors are perpendicular, no linear correlation/contribution (as discussed before).

This angular relationship is fundamental to understanding how linear models make decisions, particularly in classification tasks where the sign of the dot product determines the predicted class.

## Loss Function Mathematics in Machine Learning (Perceptron Context)

In Machine Learning, a **loss function** is a method for quantifying the "error" or "cost" associated with our model's predictions. The goal of training a model is to find the parameters (weights $\mathbf{w}$ and bias $w_0$) that minimize this loss.

### Why Do We Need a Loss Function?

Consider a linear classifier that generates a decision boundary ($\mathbf{w}^T\mathbf{x} + w_0 = 0$). After training, we want to know how well this boundary separates the data points. If a data point is on the wrong side of the boundary, our model made a mistake, and we need a way to measure the magnitude of that mistake.

### Perceptron Loss: Using Signed Distance

For a binary classification problem, we often assign labels $Y_i = +1$ for the positive class and $Y_i = -1$ for the negative class. Our model makes a prediction based on the signed output of the linear equation. We know that the signed distance of a data point $\mathbf{x}_i$ from the decision boundary is given by:

$$ d_i = \frac{\mathbf{w}^T\mathbf{x}_i + w_0}{||\mathbf{w}||} $$

This $d_i$ is positive if $\mathbf{x}_i$ is on one side of the hyperplane (e.g., the side $\mathbf{w}$ points to) and negative if it's on the other side. The magnitude of $d_i$ tells us how far the point is from the boundary.

### Penalizing Misclassified Points

A model is considered good if it places positive examples ($Y_i = +1$) on the positive side of the boundary (where $\mathbf{w}^T\mathbf{x}_i + w_0 > 0$) and negative examples ($Y_i = -1$) on the negative side (where $\mathbf{w}^T\mathbf{x}_i + w_0 < 0$).

If a point $\mathbf{x}_i$ is **correctly classified**:
-   If $Y_i = +1$, then $\mathbf{w}^T\mathbf{x}_i + w_0$ should be positive. The product $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ will be positive.
-   If $Y_i = -1$, then $\mathbf{w}^T\mathbf{x}_i + w_0$ should be negative. The product $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ will also be positive (negative times negative).

If a point $\mathbf{x}_i$ is **misclassified**:
-   If $Y_i = +1$, but $\mathbf{w}^T\mathbf{x}_i + w_0$ is negative, the point is on the wrong side. The product $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ will be negative.
-   If $Y_i = -1$, but $\mathbf{w}^T\mathbf{x}_i + w_0$ is positive, the point is on the wrong side. The product $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ will also be negative.

### Perceptron Criterion:

The Perceptron algorithm's goal is to find a separating hyperplane. It penalizes misclassified points. For a misclassified point, the product $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ will be negative. To ensure that the loss is always a positive value (which we want to minimize), we consider the negative of this product for misclassified points.

The unnormalized score for a data point $\mathbf{x}_i$ with label $Y_i$ is $S_i = Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$.

If $S_i > 0$, the point is correctly classified, and no loss is incurred by the perceptron criterion for that point.
If $S_i \le 0$, the point is misclassified, and the loss for that point is proportional to $-S_i$.

The raw score, before dividing by $||\mathbf{w}||$, is $\mathbf{w}^T\mathbf{x}_i + w_0$. If we multiply this by $Y_i$, we get $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$.

Let's define the Perceptron Loss for a single misclassified point as:

$$ L_i = -Y_i (\mathbf{w}^T\mathbf{x}_i + w_0) \quad \text{if } Y_i (\mathbf{w}^T\mathbf{x}_i + w_0) \le 0 $$

If the term $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ is already positive (correctly classified), then the loss for that point is 0.

The overall Perceptron Loss is the sum of these losses for all misclassified points:

$$ L(\mathbf{w}, w_0) = \sum_{i \in \text{misclassified}} -Y_i (\mathbf{w}^T\mathbf{x}_i + w_0) $$

This form of the loss function drives the learning algorithm to adjust $\mathbf{w}$ and $w_0$ such that all points are eventually correctly classified (i.e., $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0) > 0$ for all $i$), at which point the loss becomes zero. The term $||\mathbf{w}||$ is often omitted from the loss function itself during optimization (as it's a constant that scales the gradients) and is only crucial for calculating the *true* geometric distance.

### Gain Function: The Counterpart to Loss

A **Gain Function** (or Utility Function, Reward Function) is essentially the opposite of a Loss Function. While a loss function quantifies the penalty for incorrect predictions or model errors (which we aim to minimize), a gain function quantifies the benefit or reward for correct predictions or desired outcomes (which we aim to maximize).

**Terminology Brief:**
-   **Loss Function**: Measures how 'bad' a model is; we want to **minimize** it.
-   **Gain Function**: Measures how 'good' a model is; we want to **maximize** it.

In many cases, if you have a loss function $L$, a corresponding gain function $G$ might simply be its negative (or a transformation that makes higher values more desirable), e.g., $G = -L$. The optimization goal remains the same: find parameters that lead to the best model performance.

### Aggregate Score (Average Signed Distance for Evaluation)

We discussed that the term $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0)$ provides a score for each point, indicating if it's correctly classified and its 'distance' from the boundary (scaled by $||\mathbf{w}||$).

-   If $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0) > 0$, the prediction is right.
-   If $Y_i (\mathbf{w}^T\mathbf{x}_i + w_0) \le 0$, the prediction is wrong (or on the boundary).

To get a single aggregate measure of how good our classifier is across all data points, we can average these normalized signed distances. This aggregate score provides an overall sense of how well the model separates the classes, considering both correctness and confidence.

The **Aggregate Score** (or average signed distance) can be expressed as:

$$ \text{Aggregate Score} = \frac{1}{N} \sum_{i=1}^{N} \frac{(\mathbf{w}^T\mathbf{x}_i + w_0)}{||\mathbf{w}||} \cdot Y_i $$

Where:
-   $N$ is the total number of data points.
-   $(\mathbf{w}^T\mathbf{x}_i + w_0) / ||\mathbf{w}||$ is the signed geometric distance of point $\mathbf{x}_i$ from the decision boundary.
-   $Y_i$ is the true label ($+1$ or $-1$) for point $\mathbf{x}_i$.

**Interpretation of Aggregate Score:**
-   A **higher positive** aggregate score indicates a better classifier, meaning most points are correctly classified and lie far from the decision boundary (high confidence).
-   An aggregate score near **zero** or **negative** suggests a poor classifier, indicating many misclassifications or points very close to the boundary.

## Vector Addition, Subtraction, and Classifier Updates

### Vector Addition and Subtraction

When you add two vectors, say $\mathbf{p}$ and $\mathbf{q}$, the resulting vector $\mathbf{p} + \mathbf{q}$ will always lie *between* the original two vectors (forming the diagonal of a parallelogram defined by $\mathbf{p}$ and $\mathbf{q}$). This is a fundamental property of vector addition, often visualized using the parallelogram rule.

Similarly, when you subtract two vectors, say $\mathbf{p} - \mathbf{q}$, it's equivalent to adding $\mathbf{p}$ to the negative of $\mathbf{q}$ (i.e., $\mathbf{p} + (-\mathbf{q})$). The vector $-\mathbf{q}$ is simply $\mathbf{q}$ pointing in the exact opposite direction but with the same magnitude. The result $\mathbf{p} - \mathbf{q}$ will lie between $\mathbf{p}$ and $-\mathbf{q}$.

### Application to Perceptron Classifier Updates

In a Perceptron classifier, we have a decision boundary (a line or hyperplane) which is defined by a weight vector $\mathbf{W}$ (which is normal/perpendicular to the boundary) and a bias term $w_0$. If we assume the bias is incorporated into the weight and feature vectors, the decision boundary is given by $\mathbf{W}^T\mathbf{X} = 0$.

When a data point $\mathbf{X}_i$ is **misclassified**, the Perceptron algorithm adjusts the weight vector $\mathbf{W}$ to correctly classify that point. The update rule is typically:

$$ \mathbf{W}_{new} = \mathbf{W}_{old} + \alpha (Y_i \cdot \mathbf{X}_i) $$

Where:
- $\mathbf{W}_{old}$ is the current weight vector.
- $\mathbf{W}_{new}$ is the updated weight vector.
- $\alpha$ is the learning rate (a small positive scalar).
- $Y_i$ is the true label of the misclassified point (e.g., $+1$ or $-1$).
- $\mathbf{X}_i$ is the feature vector of the misclassified point.

Let's analyze the term $(Y_i \cdot \mathbf{X}_i)$:

-   If $Y_i = +1$ (positive class) and the point is misclassified, it means $\mathbf{W}^T\mathbf{X}_i \le 0$. To move the decision boundary so that $\mathbf{X}_i$ is correctly classified (i.e., $\mathbf{W}^T\mathbf{X}_i > 0$), we need to shift $\mathbf{W}$ in the direction of $\mathbf{X}_i$. This is achieved by **adding** $\mathbf{X}_i$ (scaled by $\alpha$) to $\mathbf{W}_{old}$: $\mathbf{W}_{new} = \mathbf{W}_{old} + \alpha \mathbf{X}_i$.

-   If $Y_i = -1$ (negative class) and the point is misclassified, it means $\mathbf{W}^T\mathbf{X}_i \ge 0$. To move the decision boundary so that $\mathbf{X}_i$ is correctly classified (i.e., $\mathbf{W}^T\mathbf{X}_i < 0$), we need to shift $\mathbf{W}$ in the direction opposite to $\mathbf{X}_i$. This is achieved by **subtracting** $\mathbf{X}_i$ (scaled by $\alpha$) from $\mathbf{W}_{old}$, which is equivalent to adding $(-\mathbf{X}_i)$: $\mathbf{W}_{new} = \mathbf{W}_{old} - \alpha \mathbf{X}_i$.

In both cases, the update rule $Y_i \cdot \mathbf{X}_i$ correctly captures the direction needed to adjust the weight vector. By adding or subtracting the misclassified point's feature vector (scaled by its label and learning rate), the normal vector $\mathbf{W}$ is nudged in a direction that helps correct the classification, effectively rotating and/or shifting the decision boundary.

### Updating the Bias Term ($w_0$)

Similar to the weight vector $\mathbf{W}$, the bias term $w_0$ also needs to be updated when a point is misclassified. The update rule for $w_0$ in the Perceptron algorithm is analogous to that for $\mathbf{W}$:

$$ w_{0,new} = w_{0,old} + \alpha (Y_i \cdot 1) $$

Here, the 'feature' corresponding to $w_0$ is implicitly a constant $1$. So, for a misclassified point $\mathbf{X}_i$ with label $Y_i$:

-   If $Y_i = +1$, $w_0$ is increased by $\alpha$.
-   If $Y_i = -1$, $w_0$ is decreased by $\alpha$.

These updates collectively shift the decision boundary to better classify the misclassified point.

## A Brief Note on the Perceptron Algorithm in Machine Learning

The Perceptron is one of the simplest linear classification algorithms and holds a significant place in the history of Artificial Intelligence and Machine Learning.

### Goal:

The primary goal of the Perceptron algorithm is to find a **linear classifier line (or hyperplane)** that can perfectly separate two classes of data points. That is, it aims to find parameters ($\mathbf{W}$ and $w_0$) such that all positive examples are on one side of the boundary and all negative examples are on the other.

### How it Works (Iterative Update):

1.  **Initialization**: Start with a randomly initialized weight vector $\mathbf{W}$ and bias term $w_0$.
2.  **Iteration**: The algorithm iterates through the training data, one point at a time.
3.  **Prediction**: For each data point $\mathbf{X}_i$, it makes a prediction based on the current $\mathbf{W}$ and $w_0$.
4.  **Update (Learning)**: If a data point is **misclassified**, the algorithm updates $\mathbf{W}$ and $w_0$ using the rules discussed above:
    -   $\mathbf{W}_{new} = \mathbf{W}_{old} + \alpha (Y_i \cdot \mathbf{X}_i)$
    -   $w_{0,new} = w_{0,old} + \alpha Y_i$
    This update moves the decision boundary slightly, nudging it closer to correctly classifying the misclassified point.
5.  **Convergence**: The process continues until all data points are correctly classified (i.e., the loss is zero), or a maximum number of iterations is reached.

### Historical Significance:

-   The Perceptron was developed by **Frank Rosenblatt in 1957**, making it one of the very first algorithms capable of learning from data.
-   It represents a single **artificial neuron**, serving as the foundational concept for much more complex **Neural Networks**.
-   Its simplicity and ability to classify linearly separable data were groundbreaking at the time, laying crucial groundwork for the field of Machine Learning.

### Understanding the Formula for `x2`

The formula `x2=(-w[0]*x1-w0)/w[1]` is used to define and plot a straight line, which in the context of our Perceptron or any linear model, represents the **decision boundary**.

#### 1. The General Equation of a Linear Decision Boundary

In a 2D space (where we have two features, `x1` and `x2`), a linear decision boundary is typically represented by the equation:

$$ w_1x_1 + w_2x_2 + w_0 = 0 $$

Where:
- $w_1$ is the weight associated with the first feature ($x_1$).
- $w_2$ is the weight associated with the second feature ($x_2$).
- $w_0$ is the bias term.

In your code, `w[0]` corresponds to $w_1$, `w[1]` corresponds to $w_2$, and `w0` is the bias.
So, the equation can be written as:

$$ \text{w[0]} \cdot x_1 + \text{w[1]} \cdot x_2 + \text{w0} = 0 $$

#### 2. Deriving `x2` from the Equation

To plot a line, it's often easiest to express one variable in terms of the other (e.g., $y = mx + c$). Here, we want to express $x_2$ in terms of $x_1$. Let's rearrange the decision boundary equation:

1.  Start with the decision boundary equation:
    $$ \text{w[0]} \cdot x_1 + \text{w[1]} \cdot x_2 + \text{w0} = 0 $$

2.  Move terms not involving $x_2$ to the right side of the equation:
    $$ \text{w[1]} \cdot x_2 = -\text{w[0]} \cdot x_1 - \text{w0} $$

3.  Divide both sides by `w[1]` to isolate $x_2$:
    $$ x_2 = \frac{-\text{w[0]} \cdot x_1 - \text{w0}}{\text{w[1]}} $$

This is exactly the formula you used: `x2=(-w[0]*x1-w0)/w[1]`.

#### 3. Why is this useful?

This derived formula allows you to:
-   **Calculate `x2` values**: For any given `x1` values (like `np.array([0,8])` in your code), you can compute the corresponding `x2` values that lie on the decision boundary.
-   **Plot the decision boundary**: Once you have a few pairs of `(x1, x2)` coordinates that lie on the line, you can use a plotting library (like `matplotlib.pyplot`) to draw the decision boundary. This visual representation helps you see how the classifier separates the data points into different classes.

```python
def perceptron_training(X,Y,w,w0,num_itr=100):
  for itr in range(num_itr):
    for i in range(X.shape[0]):
      xi = X[i]
      yi = Y[i]
      d = dist_from_hyperplane(w,w0,xi)
      pred_sign = np.sign(d)

      if pred_sign!=yi:
        w=w+(xi*yi)
        w0=w0+yi
  return w,w0
```

## Summary of Mathematical Fundamentals for Machine Learning

This notebook has explored the foundational mathematical concepts, primarily from Linear Algebra, that underpin many Machine Learning algorithms, with a focus on linear models and the Perceptron.

### 1. Mapping Functions in ML

*   **Single Feature**: `y = mx + c` (where `m` is slope, `c` is intercept)
*   **General ML Form (Multiple Features)**: `ŷ = w₀ + w₁x₁ + w₂x₂ + ... + wₙxₙ`
    *   `w₀` = bias/intercept
    *   `w₁, ..., wₙ` = weights (analogous to slope for each feature)
*   **Binary Classification**: Prediction based on the sign of `ŷ = w₀ + w₁x₁` (e.g., `ŷ > 0` for positive class, `ŷ < 0` for negative class, `ŷ = 0` on the boundary).

### 2. Vectors and Norms

*   **Vectors**: Quantities with magnitude and direction. In ML, data points ($\mathbf{X}$), feature sets, and weights ($\mathbf{W}$) are often represented as vectors.
*   **Magnitude (L2 Norm)**: Measures the size of a vector. Often used for distance.
    *   `||x||₂ = √(x₁² + x₂² + ... + xₙ²)`
*   **L1 Norm (Manhattan Norm)**: Sum of absolute values. Used in Lasso regularization for feature selection.
    *   `||x||₁ = |x₁| + |x₂| + ... + |xₙ|`
*   **Direction in ML**: Often more important than magnitude, especially when comparing similarity or relationships between vectors in feature space. Unit vectors capture pure direction.

### 3. From Linear Equations to Vector/Matrix Notation

*   **Compact Form**: `ŷ = WᵀX + w₀`
    *   Where $\mathbf{W}$ and $\mathbf{X}$ are column vectors.
*   **Dot Product**: `WᵀX = w₁x₁ + w₂x₂ + ... + wₙxₙ`
    *   Geometric interpretation: `WᵀX = ||W|| ||X|| cos(θ)`
    *   `cos(θ)` indicates alignment/similarity:
        *   `θ = 0° (cos(θ)=1)`: Same direction, strong alignment.
        *   `θ = 90° (cos(θ)=0)`: Perpendicular, no linear relationship.
        *   `θ = 180° (cos(θ)=-1)`: Opposite direction, strong opposition.
*   **Incorporating Bias ($w_0$)**: Augment feature vector ($\mathbf{X}$) with a `1` and weight vector ($\mathbf{W}$) with `w₀` to write `ŷ = W_augmentedᵀX_augmented`.

### 4. Special Vector Concepts

*   **Unit Vector**: A vector with magnitude 1. `v̂ = v / ||v||`. Captures pure direction.
*   **Projection Vector**: The component of one vector ($\mathbf{x}$) that lies in the direction of another ($\mathbf{w}$). `p = (xᵀw / ||w||²)w`
    *   Relevance: Dimensionality reduction (e.g., PCA), feature engineering, geometric understanding.
*   **Scalar Projection**: The signed magnitude of the vector projection. `comp_w x = ||x|| cos(θ) = xᵀw / ||w||`
    *   Quantifies how much of $\mathbf{x}$ aligns with $\mathbf{w}$.

### 5. Linear Decision Boundaries

*   **Equation**: `w₁x₁ + w₂x₂ + w₀ = 0` (for 2D) or `wᵀx + w₀ = 0` (general).
*   **Shifting**: Horizontal/vertical shifts are achieved by modifying the bias term `w₀`. `w₀_new = w₀ - w₁a` (horizontal) or `w₀_new = w₀ - w₂b` (vertical).
*   **Orthogonality**: The weight vector $\mathbf{w}$ is always **perpendicular (orthogonal)** to the decision boundary. It acts as the **normal vector** to the hyperplane.
*   **Distance from Origin**: `d = -w₀ / ||w||` (signed distance), or `d = |w₀| / ||w||` (magnitude).
*   **Distance of a Data Point ($\mathbf{x}₀$) from Boundary**: `d = (wᵀx₀ + w₀) / ||w||` (signed distance).
    *   `d > 0`: Point on the positive side.
    *   `d < 0`: Point on the negative side.
    *   `d = 0`: Point on the boundary.
*   **Distance Between Two Parallel Hyperplanes**: `Distance = |w₀,₁ - w₀,₂| / ||w||`.
*   **Distance Between Non-Parallel Hyperplanes**: Always `0` (because they intersect).

### 6. Angle Interpretation for Non-Perpendicular Vectors

*   **Acute Angle ($0° \le \theta < 90°$)**: `cos(θ) > 0`. Vectors generally aligned, positive contribution to prediction.
*   **Obtuse Angle ($90° < \theta \le 180°$)**: `cos(θ) < 0`. Vectors generally opposed, negative contribution to prediction.

### 7. Loss Function Mathematics (Perceptron Context)

*   **Purpose**: Quantify prediction error; used for optimization.
*   **Perceptron Loss for a Misclassified Point**: `Lᵢ = -Yᵢ (wᵀxᵢ + w₀)` (if `Yᵢ (wᵀxᵢ + w₀) \le 0`).
*   **Total Perceptron Loss**: Sum of `Lᵢ` for all misclassified points.
*   **Gain Function**: Opposite of loss, quantifies benefit (aim to maximize).
*   **Aggregate Score (Average Signed Distance)**: `(1/N) Σ ((wᵀxᵢ + w₀) / ||w||) Yᵢ`. Higher positive score = better classifier.

### 8. Perceptron Algorithm

*   **Goal**: Find a linear decision boundary to separate two classes.
*   **Iterative Update**: Adjusts `W` and `w₀` when a point is misclassified.
    *   **Weight Update**: `W_new = W_old + α (Yᵢ ⋅ Xᵢ)`
    *   **Bias Update**: `w₀_new = w₀_old + α Yᵢ`
*   **Convergence**: Continues until all points are correctly classified or max iterations reached.
*   **Historical Significance**: Developed by Frank Rosenblatt (1957), foundational for Neural Networks.
