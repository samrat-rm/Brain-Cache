#### 17. What is the result of the following expression? (★☆☆)
```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

#### Answer :
nan

False

False

nan

True

False

````python
print(0 * np.nan)
print(np.nan == np.nan)
print(np.inf > np.nan)
print(np.nan - np.nan)
print(np.nan in set([np.nan]))
print(0.3 == 3 * 0.1)
````

### Insight :

```python
print(0 * np.nan)              # nan
```
Any arithmetic involving `nan` produces `nan` because the result is undefined.

```python
print(np.nan == np.nan)        # False
```
By IEEE-754 rules, `nan` is not equal to anything, including itself.

```python
print(np.inf > np.nan)         # False
```
Any comparison with `nan` (`==`, `!=`, `<`, `>`, `<=`, `>=`) is considered undefined and returns `False` (except `!=`, which returns `True`).

```python
print(np.nan - np.nan)         # nan
```
Undefined arithmetic propagates `nan`.

```python
print(np.nan in set([np.nan])) # True
```
The set contains the exact same `np.nan` object, so membership succeeds via hash lookup and identity checks.

```python
print(0.3 == 3 * 0.1)          # False
```
Floating-point numbers are stored as binary approximations, and `3 * 0.1` evaluates to `0.30000000000000004`, not exactly `0.3`.

```python
# Preferred float comparison
np.isclose(0.3, 3 * 0.1)       # True
```
`np.isclose()` checks whether two floating-point values are equal within a small numerical tolerance.
