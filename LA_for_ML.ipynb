### Why Machine Learning Uses `w₀ + w₁x₁ + w₂x₂ + ...` Instead of `y = mx + c`

The equation

$$
y = mx + c
$$

works well for a **single input variable**. Here:

- `m` = slope (how much `y` changes when `x` changes)
- `c` = intercept (value of `y` when `x = 0`)

However, machine learning models often have **many input features**:

- House size
- Number of bedrooms
- Age of house
- Distance to city

A single slope `m` is no longer enough.

Instead, we use:

$$
\hat{y} = w_0 + w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

where:

- `w₀` = bias/intercept (same role as `c`)
- `w₁, w₂, ..., wₙ` = weights (same role as slope `m`)
- `x₁, x₂, ..., xₙ` = input features

For one feature:

$$
\hat{y} = w_0 + w_1x_1
$$

which is mathematically identical to:

$$
y = mx + c
$$

with:

$$
m = w_1,\qquad c = w_0
$$

#### Why Not Use Slope?

The concept of a single slope comes from a 2D line.

A slope is defined as:

$$
m = \frac{\Delta y}{\Delta x}
$$

and breaks down for a vertical line because:

$$
m = \frac{\Delta y}{0}
$$

(because there is no change in x $\Delta x$ is 0 )

which is undefined.

In higher dimensions (many features), there is no single "slope" anyway. Instead, each feature gets its own weight:

$$
w_1,\; w_2,\; \ldots,\; w_n
$$

This generalizes linear equations from a line to planes and hyperplanes.

## Key Idea

`y = mx + c` is a special case of linear regression with one feature.

$$
y = mx + c
\quad\Longleftrightarrow\quad
\hat{y} = w_0 + w_1x_1
$$

Machine learning uses weights because they naturally extend to any number of input features.

- If two lines are parelel then the slope of m1, m2 is equal

$m_1 = m_2$
- If two lines are perpendicular then the multiple of slopes is -1

$m_1 * m_2 = -1$

# Vectors in Machine Learning

A **vector** is a quantity that has both:

- **Magnitude** (how large it is)
- **Direction** (where it points)

> A vector is always a column. In ML, Every data point $X$ (independent feature) is a vector.

A vector starting from origin is called a **positional vector**.

For example, the vector

$$
\mathbf{v} =
\begin{bmatrix}
3 \\
4
\end{bmatrix}
$$

has magnitude:

$$
||\mathbf{v}|| = \sqrt{3^2 + 4^2} = 5
$$

and points in the direction $(3,4)$.

## What Does a Vector Mean in ML?

In machine learning, vectors usually represent **features**, **weights**, or **embeddings**.

A feature vector:

$$
\mathbf{x} =
\begin{bmatrix}
\text{Age} \\
\text{Income} \\
\text{Experience}
\end{bmatrix}
$$

describes a single data point.

A weight vector:

$$
\mathbf{w} =
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}
$$

describes how important each feature is to the prediction.

## Magnitude in ML

The magnitude tells us the overall "size" of a vector:

$$
||\mathbf{x}|| =
\sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}
$$

Large magnitude often means:

- Larger feature values
- Greater distance from the origin
- Stronger activation in some models

This is why features are often normalized before training.

## Direction in ML

Direction is often more important than magnitude.

Two vectors can have different lengths but point in nearly the same direction:

$$
\begin{bmatrix}
1 \\
2
\end{bmatrix}
\quad\text{and}\quad
\begin{bmatrix}
10 \\
20
\end{bmatrix}
$$

These represent similar patterns because they point in the same direction.

Many ML algorithms compare vectors using their direction rather than their size.

# L1 and L2 Norms in Machine Learning

A **norm** measures the size (magnitude) of a vector.

For a vector:

$$
\mathbf{x} =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

the most common norms are **L1** and **L2**.

---

## L1 Norm (Manhattan Norm / Taxicab Norm)

The L1 norm is the sum of the absolute values of all components:

$$
||\mathbf{x}||_1 =
|x_1| + |x_2| + \cdots + |x_n|
$$


Here $||\mathbf{x}||_1$ the 1 means L1 norm.


### Example

For:

$$
\mathbf{x} =
\begin{bmatrix}
3 \\
-4
\end{bmatrix}
$$

$$
||\mathbf{x}||_1
=
|3| + |-4|
=
3 + 4
=
7
$$

### Why is it called Manhattan Norm?

Imagine moving through city blocks in Manhattan:

- 3 blocks east
- 4 blocks north

Total distance traveled:

$$
3 + 4 = 7
$$

You cannot move diagonally.

### Used In

**Lasso Regression**

L1 regularization penalty:

$$
\lambda \sum |w_i|
$$

Effect:

- Pushes some weights exactly to zero
- Performs automatic feature selection
- Creates sparse models

---

## L2 Norm (Euclidean Norm)

Displacement means the shortest distance between two points which is calculated using eucledian distance.

The L2 norm is the square root of the sum of squared components:

$$
||\mathbf{x}||_2
=
\sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}
$$

### Example

For:

$$
\mathbf{x} =
\begin{bmatrix}
3 \\
4
\end{bmatrix}
$$

$$
||\mathbf{x}||_2
=
\sqrt{3^2 + 4^2}
=
\sqrt{25}
=
5
$$

### Why is it called Euclidean Norm?

It measures the straight-line distance between two points using the Pythagorean theorem.

### Used In

**Ridge Regression**

L2 regularization penalty:

$$
\lambda \sum w_i^2
$$

Effect:

- Shrinks weights toward zero
- Rarely makes weights exactly zero
- Keeps all features but reduces overfitting

---

## Quick Comparison

| Property | L1 Norm | L2 Norm |
|----------|---------|---------|
| Actual Name | Manhattan / Taxicab Norm | Euclidean Norm |
| Formula | $$\sum |x_i|$$ | $$\sqrt{\sum x_i^2}$$ |
| Example for (3,4) | 7 | 5 |
| Regularization | Lasso | Ridge |
| Feature Selection | Yes | No |
| Produces Sparse Model | Yes | No |
| Handles Correlated Features | Less Stable | More Stable |

(Regularization Lasso and Ridge comes into play later)

## Rule of Thumb

- Use **L1 (Lasso)** when you want the model to automatically remove unimportant features.
- Use **L2 (Ridge)** when you want to keep all features but prevent very large weights.
- Use **Elastic Net** when you want a combination of both.

## Vector Direction in Feature Space (Not Actual Space)

In Machine Learning, when we talk about a vector's **direction**, it's crucial to understand that we are referring to its direction within a **feature space**, not necessarily a physical, real-world direction. Each axis in this multi-dimensional space corresponds to a specific feature ($x_1, x_2, \ldots, x_n$).

-   **Feature Dominance**: If a vector has a large value along one axis and small (or zero) values along others, it means that the feature corresponding to that axis is dominant for that data point.

  *Example*:
  
   For a vector $\mathbf{v} = \begin{bmatrix} 0 \\ 0 \\ 10 \\ 0 \end{bmatrix}$, it indicates that only the third feature ($x_3$) is significant for this data point, while other features have little to no influence.

-   **Feature Relationships**: The relative values and signs of components within a vector can reveal relationships between features for a particular data point.

  *Example*: For a vector $\mathbf{v} = \begin{bmatrix} 0 \\ 1 \\ 0 \\ -1 \end{bmatrix}$,
  
  it suggests that the second feature ($x_2$) and the fourth feature ($x_4$) are important, but they have an opposing relationship. If $x_2$ has a positive contribution, $x_4$ has a negative one, implying an indirect proportionality or inverse relationship within the feature set for this specific instance.

  The **unit vector**, calculated as
  
   $$ \hat{\mathbf{v}} = \frac{\mathbf{v}}{||\mathbf{v}||} $$
   
   specifically captures this directional relationship by normalizing the vector's magnitude. It shows the pure direction in feature space, allowing us to compare the relative contributions and relationships of features independent of their overall scale.

# From Linear Equations to Vectors and Matrices

In machine learning, a prediction with one feature is written as:

$$
\hat{y} = w_1x_1 + w_0
$$

where:

- $x_1$ = input feature
- $w_1$ = weight
- $w_0$ = bias (intercept)

When we have multiple features, the equation becomes:

$$
\hat{y} = w_1x_1 + w_2x_2 + \cdots + w_nx_n + w_0
$$

Instead of writing every term separately, we group the inputs and weights into **vectors**:

$$
\mathbf{W} =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
,\qquad
\mathbf{X} =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

The prediction can now be written as a **dot product**:

$$
\hat{y} = \mathbf{W}^T\mathbf{X} + w_0
$$

where:

$$
\mathbf{W}^T\mathbf{X}
=
w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

## Moving to Matrices

For a dataset containing many training examples, we stack all feature vectors into a matrix:

$$
X =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{m1} & x_{m2} & \cdots & x_{mn}
\end{bmatrix}
$$

where:

- $m$ = number of samples
- $n$ = number of features

The weight vector remains:

$$
W =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
$$


## Key Idea

- A **scalar** is a single number.
- A **vector** is an ordered list of numbers.
- A **matrix** is a collection of vectors arranged in rows and columns.

Machine learning uses vectors and matrices because they allow predictions for thousands or millions of samples to be computed efficiently using linear algebra.

## Dot Product: Where Vectors Meet ML

The prediction equation:

$$
\hat{y} = \mathbf{w}^T\mathbf{x} + b
$$

uses a dot product. We use $W^T$ because we can't multiply $W . X$ directly due to matrix multiplication rules. When two vectors $\mathbf{w}$ and $\mathbf{x}$ meet, there is an angle $\theta$ between them. The dot product can be expressed as:

$$
\mathbf{w}^T\mathbf{x}
=
||\mathbf{w}||
\,||\mathbf{x}||
\cos(\theta)
$$

The value of $\cos(\theta)$ indicates the degree of similarity or alignment between these two vectors, giving us critical insights into their relationship:
- If $\theta = 0^{\circ}$ (vectors point in the same direction, regardless of magnitude), then $\cos(0^{\circ}) = 1$. The dot product is maximized, indicating strong alignment.
- If $\theta = 180^{\circ}$ (vectors point in opposite directions), then $\cos(180^{\circ}) = -1$. The dot product is minimized (most negative), indicating strong opposition.
- If $\theta = 90^{\circ}$ (vectors are perpendicular), then $\cos(90^{\circ}) = 0$. The dot product is zero, indicating no linear relationship or orthogonal vectors.

From this, we can also define cosine similarity, which is often used to measure the similarity between two vectors:

$$
\cos(\theta) = \frac{\mathbf{w}^T\mathbf{x}}{||\mathbf{w}|| \,||\mathbf{x}||}
$$

**Interpreting the Angle and Direction:**

When we consider the dot product $\mathbf{w}^T\mathbf{x}$ (often ignoring the bias $b$ for geometric intuition):

- The sign of $\cos(\theta)$ tells us about the angle between $\mathbf{w}$ and $\mathbf{x}$:
    - If $\cos(\theta)$ is **positive**, $\theta$ is an **acute angle** ($0^{\circ} \le \theta < 90^{\circ}$). The vectors are pointing generally in the same direction.
    - If $\cos(\theta)$ is **negative**, $\theta$ is an **obtuse angle** ($90^{\circ} < \theta \le 180^{\circ}$). The vectors are pointing generally in opposite directions.
    - If $\cos(\theta)$ is **zero**, $\theta$ is exactly $90^{\circ}$. The vectors are **perpendicular** (orthogonal).

In classification tasks, the weight vector $\mathbf{w}$ is often perpendicular to the decision boundary (the line or hyperplane that separates classes).

- If $\mathbf{w}^T\mathbf{x}$ is **positive**, the feature vector $\mathbf{x}$ lies on the same side as $\mathbf{w}$ (the 'positive' side) relative to the decision boundary.
- If $\mathbf{w}^T\mathbf{x}$ is **negative**, the feature vector $\mathbf{x}$ lies on the opposite side of $\mathbf{w}$ (the 'negative' side) relative to the decision boundary.
- If $\mathbf{w}^T\mathbf{x}$ is **zero**, the feature vector $\mathbf{x}$ lies directly on the decision boundary.

Based on $\cos(\theta)$:
- Same direction $\rightarrow$ large positive prediction
- Perpendicular $\rightarrow$ little influence
- Opposite direction $\rightarrow$ negative influence


#### Matrix Multiplication

In machine learning, predictions are often written as:

$$
\hat{y} = W^T X
$$

where:

$$
W =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
,\qquad
X =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

After transposing :

$$
W^T =
\begin{bmatrix}
w_1 & w_2 & \cdots & w_n
\end{bmatrix}
$$

The matrix multiplication rule is:

> **Number of columns in the first matrix must equal the number of rows in the second matrix.**

So:

$$
(1 \times n) \cdot (n \times 1)
=
(1 \times 1)
$$

which produces a single value:

$$
W^T X
=
w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

This is simply the **dot product** of the weight vector and feature vector.

### Example

$$
W^T =
\begin{bmatrix}
2 & 3
\end{bmatrix}
,\qquad
X =
\begin{bmatrix}
4 \\
5
\end{bmatrix}
$$

$$
W^T X
=
(2 \times 4) + (3 \times 5)
=
8 + 15
=
23
$$

Therefore:

$$
\hat{y} = 23
$$

## Key Idea

Matrix multiplication lets us compute:

$$
w_1x_1 + w_2x_2 + \cdots + w_nx_n
$$

compactly and efficiently, which is why almost all machine learning models are written using vectors and matrices rather than long equations.

---

## Incorporating the Bias Term ($w_0$) into the Dot Product

When you see the prediction equation written as:

$$ \hat{y} = \mathbf{W}^T\mathbf{X} $$

and previously it was:

$$ \hat{y} = w_0 + w_1x_1 + w_2x_2 + \cdots + w_nx_n $$

It might seem like the bias term $w_0$ has been omitted. However, for mathematical convenience and to simplify notation, the bias term is often incorporated directly into the weight vector and feature vector.

### How the Bias is Included:

To include $w_0$ in the dot product, we can modify the feature vector $\mathbf{X}$ by adding a constant value of $1$ as an additional feature, and then correspondingly modify the weight vector $\mathbf{W}$ to include $w_0$ as its first (or last) component.

Let's assume we augment the feature vector $\mathbf{X}$ as follows:

$$ \mathbf{X}_{augmented} = \begin{bmatrix} 1 \\ x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} $$

And correspondingly, we augment the weight vector $\mathbf{W}$:

$$ \mathbf{W}_{augmented} = \begin{bmatrix} w_0 \\ w_1 \\ w_2 \\ \vdots \\ w_n \end{bmatrix} $$

Now, the dot product of these augmented vectors:

$$ \mathbf{W}_{augmented}^T \mathbf{X}_{augmented} = w_0 \cdot 1 + w_1x_1 + w_2x_2 + \cdots + w_nx_n $$

Which is exactly the original linear equation, including the bias term $w_0$.

$$ \hat{y} = \mathbf{W}_{augmented}^T \mathbf{X}_{augmented} = w_0 + \sum_{i=1}^{n} w_ix_i $$

### Why do this?

This technique allows us to express the entire linear model, including the intercept, as a single matrix multiplication or dot product. This simplifies the mathematical formulation, especially when dealing with optimizations, derivatives, and implementing these models in code (where it can be more efficient to handle all parameters in one vector).

## Unit Vector

A vector that has a magnitude of 1

Convert any vector into a unit vector by dividing the values by the magnitude.

For a vector $\mathbf{v}$, its unit vector $\hat{\mathbf{v}}$ is given by:
$$\hat{\mathbf{v}} = \frac{\mathbf{v}}{||\mathbf{v}||}$$
where $||\mathbf{v}||$ is the magnitude (or L2 norm) of the vector $\mathbf{v}$.

## Projection Vector

The **projection of one vector onto another** essentially tells us how much of the first vector lies in the direction of the second vector. Imagine shining a light perpendicular to a line; the shadow cast on the line is the projection.

Let $\mathbf{x}$ and $\mathbf{w}$ be two vectors, and let $\mathbf{p}$ be the projection of $\mathbf{x}$ onto $\mathbf{w}$.

- The angle between $\mathbf{x}$ and $\mathbf{w}$ ($\theta$) is the same as the angle between $\mathbf{x}$ and its projection $\mathbf{p}$ onto $\mathbf{w}$.
- The projection vector $\mathbf{p}$ has the **same direction** as $\mathbf{w}$ but potentially a **different magnitude**.

From basic trigonometry, in a right-angled triangle formed by $\mathbf{x}$, $\mathbf{p}$, and the perpendicular from the tip of $\mathbf{x}$ to the line of $\mathbf{w}$:

(the weight vector $\mathbf{w}$ is perpendicular to the decision boundary (or prediction line). It defines the orientation of this boundary.)

$$
\cos(\theta) = \frac{\text{adjacent}}{\text{hypotenuse}} = \frac{||\mathbf{p}||}{||\mathbf{x}||}
$$

So, the magnitude of the projection is:

$$
||\mathbf{p}|| = ||\mathbf{x}|| \cos(\theta)
$$

We also know from the dot product section that:

$$
\cos(\theta) = \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{x}|| \,||\mathbf{w}||}
$$

Now, substitute this expression for $\cos(\theta)$ into the magnitude of projection formula:

$$
||\mathbf{p}|| = ||\mathbf{x}|| \left( \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{x}|| \,||\mathbf{w}||} \right)
$$

We can cancel out $||\mathbf{x}||$:

$$
||\mathbf{p}|| = \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{w}||}
$$


This gives us the magnitude of the projection of $\mathbf{x}$ onto $\mathbf{w}$.

To get the actual projection **vector** $\mathbf{p}$, we multiply its magnitude by the unit vector in the direction of $\mathbf{w}$ (which is $\frac{\mathbf{w}}{||\mathbf{w}||}$):

$$
\mathbf{p} = ||\mathbf{p}|| \left( \frac{\mathbf{w}}{||\mathbf{w}||} \right) = \left( \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{w}||} \right) \left( \frac{\mathbf{w}}{||\mathbf{w}||} \right)
$$

Thus, the projection vector $\mathbf{p}$ is:

$$
\mathbf{p} = \frac{\mathbf{x}^T\mathbf{w}}{||\mathbf{w}||^2} \mathbf{w}
$$

> In essence, to calculate the projection vector, we first find its magnitude by taking the magnitude of $\mathbf{x}$ multiplied by the cosine of the angle between $\mathbf{x}$ and $\mathbf{w}$. Then, we give it the correct direction by multiplying this magnitude by the unit vector of $\mathbf{w}$.

### Relevance in Machine Learning

Projection vectors are crucial in various ML contexts:
- **Dimensionality Reduction**: Techniques like Principal Component Analysis (PCA) project high-dimensional data onto lower-dimensional subspaces, effectively finding the 'shadow' of the data on important directions.
- **Feature Engineering**: Creating new features by projecting existing ones onto a specific direction that might be more informative for a model.
- **Geometric Understanding**: Helps visualize how data points relate to decision boundaries or other important vectors (like weight vectors in linear models).

### Clarification on the Projection Vector Formula

The projection vector $\mathbf{p}$ must be a vector that points in the same direction as $\mathbf{w}$. The formula we derived:

$$ \mathbf{p} = \frac{\mathbf{x}^T\mathbf{w}}{|\|\mathbf{w}\||^2} \mathbf{w} $$

is correct because:

1.  **Scalar Factor**: The term $\frac{\mathbf{x}^T\mathbf{w}}{|\|\mathbf{w}\||^2}$ is a **scalar**. It represents the magnitude of the projection along the direction of $\mathbf{w}$.
    *   $\mathbf{x}^T\mathbf{w}$ is a scalar (the dot product).
    *   $|\|\mathbf{w}\||^2$ is also a scalar (the squared magnitude of $\mathbf{w}$).
    *   A scalar divided by a scalar is still a scalar.

2.  **Direction**: This scalar factor is then multiplied by the vector $\mathbf{w}$. Multiplying a scalar by a vector results in a new vector that is scaled (longer or shorter) but retains the **original direction** of $\mathbf{w}$. This is precisely what a projection vector should do: lie along the direction of the vector it's projected onto.

### Why $\mathbf{p} = \mathbf{x}^T\mathbf{w}^2 / |\|\mathbf{w}\||^2$ is not correct:

The expression $\mathbf{w}^2$ (a vector squared) is not a standard mathematical operation in this context. While it could sometimes refer to an element-wise square or $|\|\mathbf{w}\||^2$ (the squared magnitude), if it were meant as an element-wise square of the vector $\mathbf{w}$, then multiplying it by the scalar $\mathbf{x}^T\mathbf{w}$ would result in a vector, but it would not correctly represent the projection along $\mathbf{w}$.

More critically, if you interpret $\mathbf{w}^2$ as $|\|\mathbf{w}\||^2$, then the formula would become:

$$ \mathbf{p} = \frac{\mathbf{x}^T\mathbf{w} \cdot |\|\mathbf{w}\||^2}{|\|\mathbf{w}\||^2} = \mathbf{x}^T\mathbf{w} $$

This result (which is a scalar) cannot be a projection **vector**. A projection is inherently a vector quantity, having both magnitude and direction.

## Projection Vector and Prediction: A Conceptual Link

At a high level, the **projection vector** $\mathbf{p}$ of an input feature vector $\mathbf{x}$ onto a weight vector $\mathbf{w}$ can be thought of as:

*   **How much of $\mathbf{x}$ lies in the direction of $\mathbf{w}$?**

This value, the magnitude of the projection $|\|\mathbf{p}\||$, effectively quantifies the 'alignment' or 'contribution' of the input $\mathbf{x}$ along the direction that the model (represented by $\mathbf{w}$) considers important. In essence, this alignment is what forms the basis of the prediction, $\hat{y}$. The dot product $\mathbf{x}^T\mathbf{w}$ (which is directly related to the projection's magnitude) is the core of how the model calculates its output based on the input features and their learned importance.

## Scalar Projection (Component)

The **scalar projection** of vector $\mathbf{x}$ onto vector $\mathbf{w}$ is a single number (a scalar) that represents the signed magnitude of the vector projection $\mathbf{p}$. It tells us 'how much' of $\mathbf{x}$ is aligned with $\mathbf{w}$, but without specifying the direction (as the vector projection does).

### Formula:

The scalar projection of $\mathbf{x}$ onto $\mathbf{w}$ is denoted as $\text{comp}_{\mathbf{w}}\mathbf{x}$ or $p_w$ and can be calculated using either:

1.  **Using the angle:**
    $$ \text{comp}_{\mathbf{w}}\mathbf{x} = |\|\mathbf{x}\|| \cos(\theta) $$

2.  **Using the dot product:**
    $$ \text{comp}_{\mathbf{w}}\mathbf{x} = \frac{\mathbf{x}^T\mathbf{w}}{|\|\mathbf{w}\||} $$

Notice that this is simply the magnitude of the vector projection $\mathbf{p}$.

### Where it's used in ML:

Scalar projection is useful when:

-   **Quantifying influence**: You want to know the extent to which one feature vector influences another direction (e.g., how much an input feature vector aligns with a specific weight vector's learned direction), without needing the full vector result.
-   **Feature extraction**: In some cases, a new feature can be derived by projecting existing features onto a particular direction and taking only the scalar component.
-   **Interpreting dot products**: The dot product $\mathbf{x}^T\mathbf{w}$ is directly proportional to the scalar projection. Understanding the scalar projection helps interpret the dot product's meaning in terms of alignment and contribution.

## Vectors Representing Words: An Intuitive Example

In Machine Learning, not just numerical data but even abstract concepts like words can be represented as vectors. These are often called **word embeddings**.

Imagine a multi-dimensional space where each dimension represents a certain characteristic or semantic property. Words that are similar in meaning or context will be positioned close to each other in this space.

Let's take your example of "King" and "man":

Consider a simplified 2-dimensional space with dimensions like `Royalty` and `Gender`.

- **Man**: Might have a vector like $\begin{bmatrix} 0.2 \\ 0.9 \end{bmatrix}$ (low royalty, high masculinity).
- **Woman**: Might have a vector like $\begin{bmatrix} 0.1 \\ -0.9 \end{bmatrix}$ (low royalty, high femininity).
- **King**: Might have a vector like $\begin{bmatrix} 0.9 \\ 0.8 \end{bmatrix}$ (high royalty, high masculinity).
- **Queen**: Might have a vector like $\begin{bmatrix} 0.8 \\ -0.7 \end{bmatrix}$ (high royalty, high femininity).

When we look at the vectors for "King" and "man":

- **King vector** = $\begin{bmatrix} 0.9 \\ 0.8 \end{bmatrix}$
- **Man vector** = $\begin{bmatrix} 0.2 \\ 0.9 \end{bmatrix}$

Even though they are not identical, they share a positive value in the 'Gender' dimension (representing masculinity) and are generally in the same quadrant of the space, indicating a conceptual similarity. They are closer to each other than "King" is to "Woman" or "Queen" is to "Man" based on gender aspects.

### How Similarity is Measured

The similarity between these word vectors is often measured using **cosine similarity**, which is derived from the dot product. As we discussed, cosine similarity measures the cosine of the angle between two vectors.

- If two vectors are very similar (point in nearly the same direction), the angle between them is small, and $\cos(\theta)$ will be close to 1.
- If two vectors are dissimilar (point in different directions), the angle is large, and $\cos(\theta)$ will be closer to 0 or even -1.

For "King" and "man", their vectors will have a high cosine similarity score because they align well in the vector space, reflecting their semantic relationship.

This demonstrates how vectors allow computers to understand and process human language by representing words as numerical entities that can be mathematically compared and manipulated.



## Fundamental Notations: Scalars and Vectors

It's important to be clear about the notation used for single numbers (scalars) and collections of numbers (vectors).

### Scalar vs. Vector Notation

-   **Scalar (Single Value)**:
    A single number, typically denoted by a lowercase letter like $x$.
    Sometimes, to indicate it's an average, it might be written with a bar over it: $\bar{x}$ (read as "x bar"). For example, $\bar{x}$ often represents the mean of a set of values.

    *Example: The temperature $x = 25^{\circ}C$, or the average score $\bar{x} = 85$.*

-   **Vector (Ordered List of Values)**:
    A collection of numbers arranged in a specific order, typically denoted by a lowercase bold letter like $\mathbf{x}$ or a letter with an arrow above it: $\vec{x}$ (read as "x vector").
    In machine learning, bold lowercase letters (e.g., $\mathbf{x}$, $\mathbf{w}$) are the most common convention for vectors.

    *Example: A 2D position vector $\mathbf{x} = \begin{bmatrix} 3 \\ 4 \end{bmatrix}$, or a 3D feature vector $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}$.*

    **Key Distinction**: While $\bar{x}$ denotes a scalar (an average), $\vec{x}$ or $\mathbf{x}$ denotes a vector (an ordered list of components).

### Set Notations: Real Numbers and Spaces

- **Real Numbers ( $\mathbb{R}$ )**:

  The symbol $\mathbb{R}$ (or just $R$) represents the set of all real numbers, which includes all rational and irrational numbers. When you see:
  $x \in \mathbb{R}$
  This means "$x$ belongs to the set of real numbers," indicating that $x$ is a single, scalar real number.

  *Example: $x = 5$, $x = -3.14$, $x = \sqrt{2}$ are all in $\mathbb{R}$.*

- **n-dimensional Real Space ($\,\mathbb{R}^n$ or $R^n$)**:

  The notation $\mathbb{R}^n$ (or $R^n$) represents an $n$-dimensional space where each dimension is a real number. When you see:
  $\mathbf{x} \in \mathbb{R}^n$
  This means "the vector $\mathbf{x}$ belongs to the $n$-dimensional real space," indicating that $\mathbf{x}$ is a vector with $n$ real-valued components.

  Example: $\mathbf{x} \in \mathbb{R}^2$ means $\mathbf{x}$ is a 2-dimensional vector with real components, like $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ where $x_1 \in \mathbb{R}$ and $x_2 \in \mathbb{R}$.


  Example: $\mathbf{x} \in \mathbb{R}^3$ means $\mathbf{x}$ is a 3-dimensional vector with real components, like $\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}$.

These notations are fundamental for precisely describing the nature and dimensionality of variables in machine learning.

## Shifting the Decision Boundary (Line/Hyperplane)

In machine learning, a linear model's decision boundary is often represented by an equation like:

$$ w_1x_1 + w_2x_2 + w_0 = 0 $$

This equation defines a line (in 2D) or a hyperplane (in higher dimensions) that separates different classes. We can shift this boundary by modifying the parameters ($w_0, w_1, w_2$).

### Horizontal Shift (along $x_1$ axis, effectively changing the intercept)

If we want to shift the line by a constant `a` along the $x_1$ axis such that the new $x_1$ becomes $x_1' = x_1 + a$, we can see its effect on the original equation:

Original point: $(x_1, x_2)$
New point for shifted line: $(x_1-a, x_2)$ (because we want the new line to pass through the old point plus 'a' shift).

Let's consider the effect of adding `a` to $x_1$ in the equation for a new point $(x_1', x_2)$ that lies on the original line. If we want the *line itself* to shift such that the old $(x_1, x_2)$ point on the old line is now $(x_1+a, x_2)$ on the new line, then we replace $x_1$ with $(x_1-a)$ in the equation to describe the new line:

$$ w_1(x_1 - a) + w_2x_2 + w_0 = 0 $$

Expanding this:

$$ w_1x_1 - w_1a + w_2x_2 + w_0 = 0 $$

Rearranging terms:

$$ w_1x_1 + w_2x_2 + (w_0 - w_1a) = 0 $$

Here, `w₁a` is a constant. So, the new bias term `w₀'` becomes:

$$ w_0' = w_0 - w_1a $$

Thus, the shifted line is:

$$ w_1x_1 + w_2x_2 + w_0' = 0 $$

This shows that **horizontal shifts (left or right)** of the decision boundary can be achieved simply by modifying the intercept (bias term) `w₀`. To move the line right, we decrease `w₀`, and to move it left, we increase `w₀` (assuming $w_1 > 0$).

### Vertical Shift (along $x_2$ axis)

Similarly, to shift the line **vertically (up or down)** by a constant `b` (i.e., $x_2' = x_2 + b$), we replace $x_2$ with $(x_2 - b)$ in the original equation:

$$ w_1x_1 + w_2(x_2 - b) + w_0 = 0 $$

Expanding this:

$$ w_1x_1 + w_2x_2 - w_2b + w_0 = 0 $$

Rearranging terms:

$$ w_1x_1 + w_2x_2 + (w_0 - w_2b) = 0 $$

Again, the shift is absorbed into a modified bias term: $w_0'' = w_0 - w_2b$.

In both cases, shifting the line (or hyperplane) in a parallel manner effectively changes only the bias term ($w_0$), while the weights ($w_1, w_2$) dictate the orientation (slope) of the boundary.

## Orthogonality and Decision Boundaries: The Normal Vector

One of the most fundamental geometric properties in linear machine learning models (like logistic regression or Support Vector Machines) is the relationship between the **weight vector** ($\mathbf{w}$) and the **decision boundary**.

### The Weight Vector as a Normal Vector

In a linear classification model, the decision boundary is defined by the equation:

$$ \mathbf{w}^T\mathbf{x} + w_0 = 0 $$

When we consider points that lie *exactly on* this boundary, the term $\mathbf{w}^T\mathbf{x}$ (if we absorb $w_0$ into $\mathbf{w}$ and $\mathbf{x}$ as discussed in "Incorporating the Bias Term") or more generally, for any two points $\mathbf{x}_1$ and $\mathbf{x}_2$ on the boundary, their difference $(\mathbf{x}_1 - \mathbf{x}_2)$ lies *within* the boundary hyperplane.

It is a crucial geometric property that the **weight vector $\mathbf{w}$ is always perpendicular (orthogonal) to the decision boundary (line or hyperplane)**. For this reason, $\mathbf{w}$ is often referred to as the **normal vector** to the decision boundary.

This means that the angle between the weight vector $\mathbf{w}$ and any vector lying *within* the decision boundary will be $90^{\circ}$.

### Mathematical Explanation

Consider the simplified case where the decision boundary passes through the origin (i.e., $w_0 = 0$). The equation becomes:

$$ \mathbf{w}^T\mathbf{x} = 0 $$

This equation means that for any point $\mathbf{x}$ that lies on the decision boundary, its dot product with the weight vector $\mathbf{w}$ is zero. As established earlier, a dot product of zero implies that the two vectors are **orthogonal** (perpendicular).

### Practical Use and Interpretation in ML

1.  **Defining Orientation**: The direction of the weight vector $\mathbf{w}$ directly determines the orientation of the decision boundary. If you rotate $\mathbf{w}$, the decision boundary also rotates.

2.  **Class Separation**: The weight vector points in the direction where the model's output values increase most rapidly. Therefore, if $\mathbf{w}$ points towards one class (e.g., the positive class), then data points on the 'other side' (the negative class) will have their feature vectors pointing in a direction generally opposite to $\mathbf{w}$.

3.  **Model Interpretation**: Understanding that $\mathbf{w}$ is orthogonal to the decision boundary helps visualize how the model separates data points. The magnitude of $\mathbf{w}$ influences the 'strength' or 'confidence' of the separation, while its direction defines the separation itself.

4.  **Distance to Boundary**: The (signed) distance of a data point $\mathbf{x}$ from the decision boundary is proportional to $\mathbf{w}^T\mathbf{x} + w_0$. This concept is central to algorithms like Support Vector Machines (SVMs), where the goal is to maximize this distance for the closest points (support vectors).
